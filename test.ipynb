```python
import plotly.graph_objects as go
import pandas as pd
```

```python
index = 6
data = pd.read_csv(f"./exp-{index}/u_v_count.txt", delimiter=";", header=None)
data = data.to_numpy().T
time_step = data[0]
u = data[1]
v = data[2]
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(x = time_step, y=u, mode='lines+markers')
)

fig.add_trace(
    go.Scatter(x = time_step, y=v, mode='lines+markers')
)
```
